```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
# Filling null values with the review rating for each category with their median...if a review rating is missing for a footwear category then it will get replaced by the median value of footwear category and not overall value.... this is good practice in Analysis

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.duplicated().any()
```

```python
# Converted all columns to lowercase and coding friendly. to make code and analysis convenient

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
# Create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# Bi-Weekly and Fortnightly are same as 14 days so changing Bi-Weekly to Fortnightly
df['frequency_of_purchases'] = df['frequency_of_purchases'].replace(to_replace=['Bi-Weekly', 'Every 3 Months'], value=['Fortnightly', 'Quarterly'])
```

```python
df['frequency_of_purchases'].unique()
```

```python
# create columns purchase frequency days

freq = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Annually': 365,
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(freq)
```

```python
df[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
```
